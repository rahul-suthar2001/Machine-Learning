```python
from sklearn.datasets import load_digits
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
digits = load_digits()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

md =LogisticRegression()
```

```python
x_train,x_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=0.2)
```

```python
md.fit(x_train,y_train)
```

```python
plt.matshow(digits.images[60])
```

```python
digits.target[60]
```

```python
md.predict([digits.data[60]])
```

```python
md.score(x_test,y_test)
```

```python
md.score(digits.data,digits.target)
```

```python
md.predict(digits.data[:5])
```

```python
digits.target[:5]
```

```python
y_predicated = md.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm =confusion_matrix(y_test,y_predicated)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(cm,annot=True)
plt.xlabel("Predicated value")
plt.ylabel("Actual value")
plt.show()
```
