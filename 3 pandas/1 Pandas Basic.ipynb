## Series

```python
marks = [30,50,100,70]
names = ["A","B","C","D"]
pd_series=pd.Series(marks)
print(pd_series)
```

```python
pd_series[2]
```

```python
pd_series=pd.Series(marks,names)
print(pd_series)
```

```python
pd_series["B"]
```

```python
marks_dic = {"A":10,"B":20,"C":30}
series = pd.Series(marks_dic)
print(series)
```

## DataFrame
