```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Graph embedded into notebook
%matplotlib inline
```

```python
x = np.arange(2,15)
y = np.array(x**2)
```

```python
print(x)
print(y)
```

```python
plt.plot(x,y)
plt.title("Square of number ")
plt.xlabel("Number")
plt.ylabel("Square")
plt.show()
```

```python
from matplotlib import style
```

```python
style.use('ggplot')
num = [1,2,3,4,5]
num2 =[12,14,16,18,20]

p = [4,5,6,7,8]
q=[20,21,22,23,24]

plt.plot(num,num2,'g',label='L1',linewidth=6)
plt.plot(p,q,'r',label='L2',linewidth=6)
plt.xlabel("num 1")
plt.ylabel("num 2")
plt.title("Random")
plt.legend()
plt.grid(True,color='c')
plt.show()
```

```python
plt.bar([1,3,5,7],[10,20,5,40],label ='Parle-G',color='b',width=0.4)
plt.bar([1.39,3.39,5.39,7.39],[5,10,12,20],label ='Tiger',color='g',width=0.4)
plt.legend()
plt.xlabel("Months")
plt.ylabel("Sells")
plt.show()
```
