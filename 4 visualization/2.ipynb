```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('ggplot')
```

# scatter plot 

```python
x =[1,2,3,4,5,6]
y =[1,4,9,16,25,36]

a=[8,9,10,11,12]
b=[13,11,10,7,5]
plt.scatter(x,y,color='b',label='Positive')
plt.scatter(a,b,color='r',label='Negative')
plt.xlabel("x,a")
plt.ylabel("y,b")
plt.title("Scatter plot")
plt.legend()
plt.show()
```

## Area plot

```python
day = np.arange(1,6)
slp = np.array([5,3,4,4,2])
eat = np.array([2,3,1,4,6])
work =  np.array([1,2,2.5,1.5,2])

plt.plot([],[],color='c',label='sleep',linewidth=5)
plt.plot([],[],color='m',label='Eat',linewidth=5)
plt.plot([],[],color='y',label='Study',linewidth=5)
plt.title("Area Plot")

plt.stackplot(day,slp,work,eat)

plt.legend()
plt.show()
```

## plot pie

```python
slices= [ 7,4,5]
activities = ['sleep','Eat','Study']
plt.pie(slices,labels=activities,colors=['c','m','y'],shadow=True,startangle=90,explode=(0,0.2,0))
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=['c','m','y'],shadow=True,startangle=90,explode=(0,0.2,0),autopct='%1.2f%%')
plt.show()
```

```python
plt.pie(slices,labels=activities,colors=['c','m','y'],startangle=90,explode=(0,0,0.2),autopct='%1.2f%%')
plt.show()
```
